```python
from IPython.display import clear_output

def display_board(board):
    
    if board == []:
        print('This is how the Indexes on board looks like:')
        print(' ')
        print('     |     |     ')
        print(f'  7  |  8  |  9  ')
        print('     |     |     ')
        print('-----------------')
        print('     |     |     ')
        print(f'  4  |  5  |  6  ')
        print('     |     |     ')
        print('-----------------')
        print('     |     |     ')
        print(f'  1  |  2  |  3  ')
        print('     |     |     ')
    
    else:
        clear_output(wait=True)
        print('     |     |     ')
        print(f'  {board[7]}  |  {board[8]}  |  {board[9]}  ')
        print('     |     |     ')
        print('-----------------')
        print('     |     |     ')
        print(f'  {board[4]}  |  {board[5]}  |  {board[6]}  ')
        print('     |     |     ')
        print('-----------------')
        print('     |     |     ')
        print(f'  {board[1]}  |  {board[2]}  |  {board[3]}  ')
        print('     |     |     ')
```

```python
def player_input1():
    
    p_choise = ['x', 'o', 'X', 'O']
    player_index = 'WRONG'
    while player_index not in p_choise:
        
        player_index = input('Player 1, You would like to be X or O: ')
        
        if player_index not in p_choise:
            print('Please Try again (Type only X or O).')
        else:
            if player_index == 'x':
                player_index = 'X'
            elif player_index == 'o':
                player_index = 'O'
            else:
                pass
            
    return player_index
```

```python
def player1_input_index(player_1_choice, board):
    
    
    print('Hello Player 1, Your Turn')
    player_choice = '20'
    
    while int(player_choice) not in range(1,10):
        
        player_choice = input(f'Please enter the index to put your {player_1_choice}: ')
        
        if int(player_choice) not in range(1,10):
            print('Error, please choose an index between (1-9)')
        else:
            if board[int(player_choice)] != ' ':
                print('Invalid index, index already taken, please try again.')
                player_choice = '20'
                
            else:
                pass
        
    return int(player_choice)
```

```python
def player2_input_index(player_2_choice, board):
    
    player_choice = '20'
    print('Hello Player 2, Your Turn')
    
    while int(player_choice) not in range(1,10):
        
        player_choice = input(f'Please enter the index to put your {player_2_choice}: ')
        
        if int(player_choice) not in range(1,10):
            print('Error, please choose an index between (1-9)')
        else:
            if board[int(player_choice)] != ' ':
                print('Invalid index, index already taken, please try again.')
                player_choice = '20'
                
            else:
                pass
        
    return int(player_choice)
```

```python
def checkk_win(board):
    check_win = False
    if (board[1] == board[2] == board[3] != ' ') or (board[4] == board[5] == board[6] != ' ') or (board[7] == board[8] == board[9] != ' ') or (board[1] == board[4] == board[7] != ' ') or (board[2] == board[5] == board[8] != ' ') or (board[3] == board[6] == board[9] != ' ') or (board[1] == board[5] == board[9] != ' ') or (board[3] == board[5] == board[7] != ' '):
        return ('Win')
        
    elif (' ' not in board):
        return ('Draw')
    
    else:
        
        return ('Not ended')
```

```python
def playy_again():
    
    play_again = 'WRONG'
    while play_again not in ['Y', 'N']:
        play_again = input('Would you like to play again? (Y or N)')
        if play_again not in ['Y', 'N']:
            print('only Y or N is allowed, TRY AGAIN')
            
    if play_again == 'Y':
        return True
    else:
        return False
```

```python
 # Game
play_again = True

while play_again:
    
    play_again = True
    check_win = 'Not ended'
    display_board([])
    board = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    player_1_choice = player_input1()

    if player_1_choice == 'X':
        player_2_choice = 'O'
    else:
        player_2_choice = 'X'

    while check_win == 'Not ended':
        
        display_board(board)
        player_1_index = player1_input_index(player_1_choice, board)
        board[player_1_index] = player_1_choice
        
        check_win = checkk_win(board)
        if check_win == 'Not ended':
            display_board(board)
            player_2_index = player2_input_index(player_2_choice, board)
            board[player_2_index] = player_2_choice
            check_win = checkk_win(board)
            if check_win == 'Draw':
                display_board(board)
                print ("it's a DRAW!")
                play_again = playy_again()
            elif check_win == 'Win':
                display_board(board)
                print ('Player 2 WON! congratulations')
                play_again = playy_again()
        elif check_win == 'Draw':
            display_board(board)
            print ("it's a DRAW!")
            play_again = playy_again()
        else:
            display_board(board)
            print ('Player 1 WON! congratulations')
            play_again = playy_again()
        
```
