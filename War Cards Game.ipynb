```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card():
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        
        return (f'{self.rank} of {self.suit}')
```

```python
class Deck():
    
    def __init__(self):
        
        self.deck_cards = []
        
        for suit in suits:
            for rank in ranks:
                self.deck_cards.append(Card(suit,rank))
                
    def shuffle(self):
        
        return random.shuffle(self.deck_cards)
    
    def remove_one(self):
        
        return self.deck_cards.pop(0)
```

```python
class Player():
    
    def __init__(self, name):
        
        self.name = name
        self.own_cards = []
        
    def insert_card(self, insert):
        
        if type(insert) == type([]):
            self.own_cards.extend(insert)
        else:
            self.own_cards.append(insert)
            
    def show_and_remove(self):
        
        return self.own_cards.pop(0)
```

```python
# Game Setup:

player_one = Player('Player One')
player_two = Player('Player Two')
game_deck = Deck()
game_deck.shuffle()

# Splitting the Shuffled deck for both players

for x in range(26):
    
    player_one.insert_card(game_deck.remove_one())
    player_two.insert_card(game_deck.remove_one())
```

```python
# Game Progress

game_on = True
collect_deck = []
rround = 0
print('Game Started:')
while game_on:
    
    # Win Check
    
    if len(player_one.own_cards) == 0:
        print('Player Two Won!!!')
        print('Player One Has No Cards')
        game_one = False
        break
        
    elif len(player_two.own_cards) == 0:
        print('Player One Won!!!')
        print('Player Two Has No Cards')
        game_one = False
        break
    
    rround += 1
    print(f'Round Number {rround}')
    print(f"Player's One card is: {player_one.own_cards[0]} /n Player's Two card is: {player_two.own_cards[0]}")
    
    # WAR - Check and Operation
    
    if player_one.own_cards[0].value == player_two.own_cards[0].value:
        print('WAR begins!')
        war_on = True
        while war_on:
            
            # If a Player has less then 4 cards at the start of a War he loses
            
            if len(player_one.own_cards) < 5:
                print('Player Two Won!!!')
                print('Player One Has Not enough cards for WAR')
                game_on = False
                break
            elif len(player_two.own_cards) < 5:
                print('Player One Won!!!')
                print('Player Two Has Not enough cards for WAR')
                game_on = False
                break
            collect_deck.extend([player_one.show_and_remove(), player_two.show_and_remove()])
            for x in range(3):
                collect_deck.append(player_one.show_and_remove())
                collect_deck.append(player_two.show_and_remove())
            print(f"Player's One card is: {player_one.own_cards[0]} & Player's Two card is: {player_two.own_cards[0]}")
            
            if player_one.own_cards[0].value == player_two.own_cards[0].value:
                print('WAR starts Again')
                continue
            elif player_one.own_cards[0].value > player_two.own_cards[0].value:
                collect_deck.extend([player_one.show_and_remove(), player_two.show_and_remove()])
                print(f'Player One Won the War and Collects: {len(collect_deck)}')
                player_one.insert_card(collect_deck)
                collect_deck = []
                war_on = False
            else:
                collect_deck.extend([player_one.show_and_remove(), player_two.show_and_remove()])
                print(f'Player Two Won the War and Collects: {len(collect_deck)} Cards')
                player_two.insert_card(collect_deck)
                collect_deck = []
                war_on = False
    
    
    # Not the Same Card --> Normal Trade Check
            
    elif player_one.own_cards[0].value > player_two.own_cards[0].value:
        collect_deck.extend([player_one.show_and_remove(), player_two.show_and_remove()])
        print(f'Player One Won the Trade and Collects: {len(collect_deck)}')
        player_one.insert_card(collect_deck)
        collect_deck = []
    else:
        collect_deck.extend([player_one.show_and_remove(), player_two.show_and_remove()])
        print(f'Player Two Won the Trade and Collects: {len(collect_deck)}')
        player_two.insert_card(collect_deck)
        collect_deck = []
```
